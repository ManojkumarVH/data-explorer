```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\manoj\Desktop\exam\bank.csv')
data.head()
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=data['age'])
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['balance'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Balance')
plt.xlabel('Balance')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
labels = ['Married', 'Single', 'Divorced']
sizes = data['marital'].value_counts()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Marital Status Distribution')
plt.axis('equal')
plt.show()
```

```python
sns.pairplot(data, hue='deposit')
plt.suptitle('Pairplot with Hue for Deposit', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']

# Plot countplots for each categorical variable with hue for 'deposit'
plt.figure(figsize=(18, 12))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(3, 3, i)
    sns.countplot(data=data, x=col, hue='deposit')
    plt.title(f'{col.capitalize()} Distribution by Deposit')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
