```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv(r'C:\Users\manoj\Desktop\exam\IPL_Data.csv')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
columns_to_drop = ['Url', 'Full Name', 'National Side', 'Sport', 'R/O', '3s', '5s', 'Mtc']
df = data.drop(columns=columns_to_drop, axis=1)
df.head()
df.copy()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
sorted_data = df.sort_values(by='RunsScored', ascending=False)
top_scorer = sorted_data.iloc[0]
print(top_scorer[['Name', 'Team', 'RunsScored']])
```

```python
columns_with_missing = ['Born', 'Age', 'Batting Style', 'Bowling', 'MatchPlayed', 'InningsBatted',
                         'NotOuts', 'RunsScored', 'HighestInnScore', '100s', '50s', '4s', '6s',
                         'BattingAVG', 'BattingS/R', 'CatchesTaken', 'StumpingsMade', 'Ducks',
                         'InningsBowled', 'Overs', 'Maidens', 'RunsConceded', 'Wickets', 'Best',
                         'BowlingAVG', 'EconomyRate', 'S/R']

for column in columns_with_missing:
    if df[column].dtype == 'float64':
        df[column].fillna(df[column].mean(), inplace=True)
    else:
        df[column].fillna(df[column].mode()[0], inplace=True)

df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['RunsScored'] ,bins=30, kde=True)
plt.title('Distribution of RunsScored')
plt.xlabel('RunsScored')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='RunsScored', y='Wickets', data=df)
plt.title('RunsScored vs. Wickets')
plt.xlabel('RunsScored')
plt.ylabel('Wickets')
plt.show()
```

```python
sns.pairplot(df[['RunsScored', 'Wickets', 'BattingAVG', 'BowlingAVG']])
plt.suptitle('Pair Plot of RunsScored, Wickets, BattingAVG, and BowlingAVG', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='RunsScored', data=df)
plt.title('Box Plot of RunsScored')
plt.show()
```

```python
corr_matrix = df[['RunsScored', 'Wickets', 'BattingAVG', 'BowlingAVG']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Team', data=df, palette='YlOrRd')
plt.title('Distribution of Players Across Teams')
plt.xlabel('Team')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
batting_style_counts = df['Batting Style'].value_counts()
colors = sns.color_palette('YlGnBu', len(batting_style_counts))
batting_style_counts.plot.pie(autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Distribution of Batting Styles')
plt.ylabel('')
plt.show()
```

```python
# Bar chart comparing 'RunsScored' for top players
top_players = df.nlargest(5, 'RunsScored')  # Adjust the number of players as needed
plt.figure(figsize=(12, 6))
sns.barplot(x='Name', y='RunsScored', data=top_players, palette='inferno')
plt.title('Comparison of RunsScored for Top Players')
plt.xlabel('Player Name')
plt.ylabel('RunsScored')
plt.show()
```
